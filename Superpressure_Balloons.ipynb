## U.S. Standard Atmosphere 1976

```python
P0 = 101325.0            # Pa
T0 = 15.0                # °C
A0 = 340.294             # m/s
G0 = 9.80665             # m/s^2

Ttp = -56.5              # °C
Tsp = -2.5               # °C
Tmp = -86.28             # °C
a = -0.0065              # °C/m
b = 0.001                # °C/m
c = 0.0028               # °C/m
d = -0.0028              # °C/m
e = -0.002               # °C/m
n = 5.2561
Rair = 286.9             # J/(kg*K)
Rh = 4124                # J/(kg*K)
Rhe = 2077               # J/(kg*K)

maxAlt = 91000
H = np.arange(0,maxAlt,1) # m
T = np.zeros(maxAlt)      # °C
P = np.zeros(maxAlt)      # Pa
Da = np.zeros(maxAlt)     # kg/m^3
Dh = np.zeros(maxAlt)     # kg/m^3
Dhe = np.zeros(maxAlt)    # kg/m^3
G = np.zeros(maxAlt)      # m/s^2

for g in range(0,maxAlt):
    if g == 0:
        G[g] = G0
    else:
        G[g] = G0 * (6371008.8/(6371008.8 + H[g]))**2

for i in range(11000):
    if i == 0:
        T[i] = T0
        P[i] = P0
    else:
        T[i] = T[i-1] + a
        P[i] = P0 * ((T[i] + 273.15) / (T0 + 273.15))**(-G[0] / a / Rair)

for i in range(11000,20000):
    T[i] = Ttp
    P[i] = P[10999] * np.exp(-G[0] / Rair / (Ttp + 273.15) * (H[i] - 10999))

for i in range(20000,32000):
    T[i] = T[i-1] + b
    P[i] = P[19999] * ((T[i] + 273.15) / (Ttp + 273.15))**(-G[0] / b / Rair)

for i in range(32000,47000):
    T[i] = T[i-1] + c
    P[i] = P[31999] * ((T[i] + 273.15) / (T[31999] + 273.15))**(-G[0] / c / Rair)

for i in range(47000,51000):
    T[i] = Tsp
    P[i] = P[46999] * np.exp(-G[0] / Rair / (Tsp + 273.15) * (H[i] - 46999))

for i in range(51000,71000):
    T[i] = T[i-1] + d
    P[i] = P[50999] * ((T[i] + 273.15) / (T[50999] + 273.15))**(-G[0] / d / Rair)

for i in range(71000,84852):
    T[i] = T[i-1] + e
    P[i] = P[70999] * ((T[i] + 273.15) / (T[70999] + 273.15))**(-G[0] / e / Rair)

for i in range(84852,91000):
    T[i] = Tmp
    P[i] = P[84851] * np.exp(-G[0] / Rair / (Tmp + 273.15) * (H[i] - 84851))

for d in range(0,maxAlt):
    Da[d] = P[d] / (Rair * (T[d] + 273.15))
    Dh[d] = P[d] / (Rh * (T[d] + 273.15))
    Dhe[d] = P[d] / (Rhe * (T[d] + 273.15))

plt.plot(T, H)
plt.title('Temperature')
plt.xlabel('Temperature (C)')
plt.ylabel('Altitude (m)')
plt.grid()
plt.show()

plt.plot(P, H)
plt.title('Pressure')
plt.xlabel('Pressure (Pa)')
plt.ylabel('Altitude (m)')
plt.grid()
plt.show()

plt.plot(Da, H, 'b', label='Air')
plt.plot(Dhe, H, 'g', label='Helium')
plt.plot(Dh, H, 'r', label='Hydrogen')
plt.title('Density')
plt.xlabel('Density (kg/m^3)')
plt.ylabel('Altitude (m)')
plt.legend()
#plt.axis([0,0.2,0,91000])
plt.grid()
plt.show()

plt.plot(G, H)
plt.title('Gravity')
plt.xlabel('Gravity (m/s^2)')
plt.ylabel('Altitude (m)')
plt.grid()
plt.show()
```

## Balloon Trajectory

```python
EnvelopeVolume = 1.484       # m^3
EnvelopeMass = 0.427         # kg
PayloadMass = 0.0354         # kg
FreeLift = 0.011             # kg
SuperTemperature = 0         # °C
Gamma = 1.35                 # 
EnvelopeThickness = 0.000050 # m

GasConstant = 8.314462       # J/molK
Rspecific_Air = 286.9        # J/kgK
MolarMass_Air = 28.97        # g/mol
Density_Air = 1.22565        # kg/m^3 (0m)
Rspecific_H = 4124           # J/kgK
MolarMass_H = 2.016          # g/mol
Density_H = 0.085266675      # kg/m^3 (0m)

Cd = 0.47                    # drag coefficient
datapoints = 20000           # s (for small free lifts more datapoints necessary)
```

```python
# original model

gas_volume = np.zeros(datapoints)
weight = np.zeros(datapoints)
buoyancy = np.zeros(datapoints)
crossSectionArea = np.zeros(datapoints)
ascentRate = np.zeros(datapoints)
superpressure = np.zeros(datapoints)
freelift = np.zeros(datapoints)
trajectory1 = np.zeros(datapoints)
trajectory2 = np.zeros(datapoints)
trajectory3 = np.zeros(datapoints)
trajectory4 = np.zeros(datapoints)
EstimatedThickness = 0.0
CircumStress = 0.0
LongStress = 0.0

def modelTrajectory(output, points, bVolume, bMass, pMass, fLift, printOutput, superTemp, stretch):
    
    global EstimatedThickness
    global CircumStress
    global LongStress
    
    FullEnvelopeAltitude = 0.0
    timeToFullEnvelope = 0.0
    
    Area = 2 * np.pi * (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) ** 2
    
    bVolume = bVolume * stretch
    
    EstimatedThickness = Area * EnvelopeThickness / (2 * np.pi * \
                        (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) ** 2)
    
    RequiredLift = (bMass + pMass + fLift)

    RequiredMass_H = (RequiredLift / (Da[0] - Dh[0]) * P[0]) \
                   / (GasConstant * (T[0] + 273.15)) * (MolarMass_H / 1000)

    RequiredVolume_H = RequiredMass_H / Density_H
    
    for i in range(points):
        
        if i == 0:
            gas_volume[i] = RequiredMass_H * Rh * (T[0] + 273.15 + superTemp) / P[0]
            crossSectionArea[i] = ((gas_volume[i] / np.pi / 4 * 3)**(0.33333))**2 * np.pi
            superpressure[i] = 0
            weight[i] = 0
            buoyancy[i] = 0
            ascentRate[i] = 0
            output[i] = 0
        else:
            pressure = P[int(output[i-1])]
            temperature = T[int(output[i-1])]
            air_density = pressure / (Rair * (temperature + 273.15))
            gravity = G0 * (6371008.8/(6371008.8 + output[i-1]))**2

            gas_volume[i] = RequiredMass_H * Rh * (temperature + 273.15 + superTemp) / pressure

            if gas_volume[i] >= bVolume:
                gas_volume[i] = bVolume
            else:
                FullEnvelopeAltitude = output[i-1]
                timeToFullEnvelope = i

            superpressure[i] = RequiredMass_H * Rh * (temperature + 273.15 + superTemp) \
                             / gas_volume[i] - pressure
            
            crossSectionArea[i] = ((gas_volume[i] / np.pi / 4 * 3)**(0.33333))**2 * np.pi
            weight[i] = gravity * (bMass + pMass + RequiredMass_H)
            buoyancy[i] = air_density * gravity * gas_volume[i]
            freelift[i] = (buoyancy[i] - weight[i]) / gravity

            ascentRate[i] = np.sqrt((2 * abs((buoyancy[i] - weight[i]))) \
                          / (air_density * Cd * crossSectionArea[i]))
            
            if weight[i] > buoyancy[i]:
                ascentRate[i] = ascentRate[i] * (-1)

            output[i] = output[i-1] + ascentRate[i]
            
    CircumStress = (superpressure[points-1] * (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) \
                    / 1.311) / EstimatedThickness
    LongStress = (superpressure[points-1] * (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) \
                    * 0.457) / (2 * EstimatedThickness)
    
    if printOutput == True:
        print ""
        print "INPUT: {0:.3f}m^3 {1:.3f}kg {2:.3f}kg {3:.3f}kg Γ{4:.2}" \
              .format(bVolume, bMass, pMass, fLift, stretch)
        print "Full Envelope Altitude:\t{0:.0f}\tm".format(FullEnvelopeAltitude)
        print "Float Altitude:\t\t{0:.0f}\tm".format(output[points-1])
        print "Float Superpressure:\t{0:.0f}\tPa" \
               .format(superpressure[points-1])
        print "Float Supertemperature:\t{0:.2f}\t°C".format(superTemp)
        print "Initial Ascent Rate:\t{0:.2f}\tm/s".format(ascentRate[1])
        print "Time to Full Envelope:\t{0:.0f}\tmin".format(timeToFullEnvelope/60)
        print "Required Lift:\t\t{0:.5f}\tkg".format(RequiredLift)
        print "Required H Mass:\t{0:.5f}\tkg".format(RequiredMass_H)
        print "Required H Volume:\t{0:.5f}\tm^3".format(RequiredVolume_H)
        print "Estimated Thickness:\t{0:.2f}\tum".format(EstimatedThickness*1000000)
        print "Circumferential Stress:\t{0:.1f}\tMPa".format(CircumStress/1000000)
        print "Longitudinal Stress:\t{0:.1f}\tMPa".format(LongStress/1000000)
```

```python
# extended model (included interpolated supertemperature)

gas_volumeST = np.zeros(datapoints)
weightST = np.zeros(datapoints)
buoyancyST = np.zeros(datapoints)
crossSectionAreaST = np.zeros(datapoints)
ascentRateST = np.zeros(datapoints)
superpressureST = np.zeros(datapoints)
freeliftST = np.zeros(datapoints)
trajectory1ST = np.zeros(datapoints)
EstimatedThicknessST = 0.0
CircumStressST = 0.0
LongStressST = 0.0

altST = [0, 3000, 6000, 9000, 12000, 16000, 24000, 50000] # meters (last value limit)
tempST = [0, 8, 10, 14, 16, 19, 20, 20] # °C (last value limit)

def modelTrajectoryST(output, points, bVolume, bMass, pMass, fLift, printOutput, stretch):
    
    global EstimatedThicknessST
    global CircumStressST
    global LongStressST
    global FullEnvelopeAltitude
    global superTemp
    
    FullEnvelopeAltitude = 0.0
    timeToFullEnvelope = 0.0
    superTemp = 0.0
    
    Area = 2 * np.pi * (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) ** 2
    
    bVolume = bVolume * stretch
    
    EstimatedThicknessST = Area * EnvelopeThickness / (2 * np.pi * \
                        (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) ** 2)
    
    RequiredLift = (bMass + pMass + fLift)

    RequiredMass_H = (RequiredLift / (Da[0] - Dh[0]) * P[0]) \
                   / (GasConstant * (T[0] + 273.15)) * (MolarMass_H / 1000)

    RequiredVolume_H = RequiredMass_H / Density_H
    
    for i in range(points):
        
        if i == 0:
            gas_volumeST[i] = RequiredMass_H * Rh * (T[0] + 273.15) / P[0]
            crossSectionAreaST[i] = ((gas_volumeST[i] / np.pi / 4 * 3)**(0.33333))**2 * np.pi
            superpressureST[i] = 0
            weightST[i] = 0
            buoyancyST[i] = 0
            ascentRateST[i] = 0
            output[i] = 0
        else:
            for z in range(1,8):
                if output[i-1] < altST[z]:
                    superTemp = (((output[i-1] - altST[z-1]) / (altST[z] - altST[z-1])) * \
                                (tempST[z] - tempST[z-1])) + tempST[z-1]
                    break
            
            pressure = P[int(output[i-1])]
            temperature = T[int(output[i-1])]
            air_density = pressure / (Rair * (temperature + 273.15))
            gravity = G0 * (6371008.8/(6371008.8 + output[i-1]))**2

            gas_volumeST[i] = RequiredMass_H * Rh * (temperature + 273.15 + superTemp) / pressure

            if gas_volumeST[i] >= bVolume:
                gas_volumeST[i] = bVolume
            else:
                FullEnvelopeAltitude = output[i-1]
                timeToFullEnvelope = i

            superpressureST[i] = RequiredMass_H * Rh * (temperature + 273.15 + superTemp) \
                             / gas_volumeST[i] - pressure
            
            crossSectionAreaST[i] = ((gas_volumeST[i] / np.pi / 4 * 3)**(0.33333))**2 * np.pi
            weightST[i] = gravity * (bMass + pMass + RequiredMass_H)
            buoyancyST[i] = air_density * gravity * gas_volumeST[i]
            freeliftST[i] = (buoyancyST[i] - weightST[i]) / gravity

            ascentRateST[i] = np.sqrt((2 * abs((buoyancyST[i] - weightST[i]))) \
                          / (air_density * Cd * crossSectionAreaST[i]))
            
            if weightST[i] > buoyancyST[i]:
                ascentRateST[i] = ascentRateST[i] * (-1)

            output[i] = output[i-1] + ascentRateST[i]
            
    CircumStressST = (superpressureST[points-1] * (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) \
                    / 1.311) / EstimatedThicknessST
    LongStressST = (superpressureST[points-1] * (((5.16 * bVolume) / (2 * np.pi)) ** (1. / 3)) \
                    * 0.457) / (2 * EstimatedThicknessST)
    
    if printOutput == True:
        print ""
        print "INPUT: {0:.3f}m^3 {1:.3f}kg {2:.3f}kg {3:.3f}kg Γ{4:.2}" \
              .format(bVolume, bMass, pMass, fLift, stretch)
        print "Full Envelope Altitude:\t{0:.0f}\tm".format(FullEnvelopeAltitude)
        print "Float Altitude:\t\t{0:.0f}\tm".format(output[points-1])
        print "Float Superpressure:\t{0:.0f}\tPa" \
               .format(superpressureST[points-1])
        print "Float Supertemperature:\t{0:.2f}\t°C".format(superTemp)
        print "Initial Ascent Rate:\t{0:.2f}\tm/s".format(ascentRateST[1])
        print "Time to Full Envelope:\t{0:.0f}\tmin".format(timeToFullEnvelope/60)
        print "Required Lift:\t\t{0:.5f}\tkg".format(RequiredLift)
        print "Required H Mass:\t{0:.5f}\tkg".format(RequiredMass_H)
        print "Required H Volume:\t{0:.5f}\tm^3".format(RequiredVolume_H)
        print "Estimated Thickness:\t{0:.2f}\tum".format(EstimatedThicknessST*1000000)
        print "Circumferential Stress:\t{0:.1f}\tMPa".format(CircumStressST/1000000)
        print "Longitudinal Stress:\t{0:.1f}\tMPa".format(LongStressST/1000000)
```
